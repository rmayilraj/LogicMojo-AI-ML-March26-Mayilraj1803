Q1. Understanding Feature Distributions (Iris Dataset)

Context: Before training any ML model, we analyze feature distributions.

Dataset: Load Iris dataset from sklearn

Tasks:

1. Extract feature matrix as NumPy array
2. Compute:
Mean
Median
Standard deviation
Variance (for each feature)
3. Identify:
Which feature has highest variability and why it matters in ML
4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
from sklearn.datasets import load_iris
import numpy as np
df = load_iris()

#Extract feature matrix as NumPy array
print('--- Extract feature matrix as NumPy array ---')
feature_arr = df.data
#printing only top 5 rows for space complexity
print(feature_arr[:5])

#Compute: Mean, Median, Standard deviation, Variance (for each feature)
mean = np.mean(feature_arr, axis = 0)
median = np.median(feature_arr, axis = 0)
std_dev = np.std(feature_arr, axis = 0)
variance = np.var(feature_arr, axis = 0)

print('\n--- Compute: Mean, Median, Standard deviation, Variance (for each feature) ---')
print(f"| {'--------------------':<20} | {'----------':<10} | {'----------':<10} | {'----------':<10} | {'----------':<10} |") 
print(f"| {'Feature Name':<20} | {'Mean':<10} | {'Median':<10} | {'Std.Dev':<10} | {'Variance':<10} |")
print(f"| {'--------------------':<20} | {'----------':<10} | {'----------':<10} | {'----------':<10} | {'----------':<10} |")
for index, name in enumerate(df.feature_names):
    print(f'| {name:<20} | {mean[index]:10.6f} | {median[index]:10.6f} | {std_dev[index]:10.6f} | {variance[index]:10.6f} |')
print(f"| {'--------------------':<20} | {'----------':<10} | {'----------':<10} | {'----------':<10} | {'----------':<10} |")

print('\n--- 3. Identify: Which feature has highest variability and why it matters in ML ---')
index = np.argmax(variance)
print(f'{df.feature_names[index]} has the highest variance : {variance[index]:1.6f}')
#To Do
print('To Do : Why highest variablity matters in ML : \n'  
     '         high variance means lot of information and gives the predictive power \n' \
     '         will help in classification. \n'\
     '         Low variance can be removed from feature because all values almost same. add no value to ML')

print('\n---  Convert any one feature into shape (n,1) and explain why ML models expect this format --')
petal_len_arr = feature_arr[:, 2].reshape(-1, 1)
#printing only top 5 rows for space complexity
print(petal_len_arr[:5])
#To Do
print('To Do : Why ML models expects feature in (n,1) or (-1,1) shape : \n'\
    '       ML model always expects data like table (2D)')
```

Q3. Feature Scaling using NumPy (Very Important)

Context: Most ML models require normalized data.

Dataset: Iris (NumPy array)

Tasks:

$$
X' = \frac{X - \mu}{\sigma}
$$
	​2.

```python
from sklearn.datasets import load_iris

df = load_iris()
feature_iris = df.data
mean = np.mean(feature_iris, axis = 0)
std_dev = np.std(feature_iris, axis = 0)

feature_iris_copy = (feature_iris - mean) / std_dev
print('--- Apply standardization ---')
print('Top 5 rows after applying standardization')
print(feature_iris_copy[:5])

print('\n--- Verify: Mean ≈ 0, Std ≈ one ---')
print(f"Mean after Standardization: {np.mean(feature_iris_copy, axis = 0)}")
print(f"Standard deviation after Standardization : {np.std(feature_iris_copy, axis = 0)}")
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
import seaborn as sns

df = sns.load_dataset('titanic')
df.head()

def age_group(age):
    if age < 18:
        return 'child'
    elif age >= 60:
        return 'senior'
    else:
        return 'adult'

df['age_group'] = df['age'].apply(age_group)
survival_rate = df.groupby('age_group')['survived'].sum()*100/df.groupby('age_group')['survived'].count()
print(survival_rate)
print('age group child has the highest survival rate')
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

np.random.seed(42)
errors = np.random.normal(loc=0, scale=1.0, size=1000)
mean_val = np.mean(errors)
std_val = np.std(errors)

lower_bound = mean_val - 2 * std_val
upper_bound = mean_val + 2 * std_val
outliers = errors[(errors < lower_bound) | (errors > upper_bound)]
print('Mean : ', mean_val)
print('Std.dev : ', std_val)
print('Outlier range : ', lower_bound, ' : ', upper_bound)
print('No.of outliers found : ', len(outliers))
```
