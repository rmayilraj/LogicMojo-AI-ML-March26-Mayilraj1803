```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
print("---- Identify and display all transaction values that appear more than once. ----")
items =  list(set(item for item in transactions if transactions.count(item) > 1))
print(items)

print("\n---- Detect transactions greater than ₹5000 as these may require fraud monitoring. ----")
items =  list(item for item in transactions if item <= 5000)
print(items)

print("\n---- Calculate the total spending for the day. ----")
print(sum(transactions))

print("\n---- Create a new list containing only unique transaction values. ----")
print(list(set(transactions)))

print("\n---- Sort the unique transaction list in descending order to view the largest transactions first. ----")
print(sorted(list(set(transactions)), reverse = "true"))
```

```python
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]

print("\n---- Iterates through the dataset and removes all invalid age values. ----")
print("---- Stores only valid ages in a new list. ----")
ages_valid = []
for item in ages:
    if (item == None) or (int(item) <= 0): 
        continue
    ages_valid.append(item)
print(ages_valid)

print("\n---- Calculates the average age of valid customers. ----")
print(sum(ages_valid) / len(ages_valid))

print("\n---- Displays the list of customers whose age is greater than 30. ----")
print(list(item for item in ages_valid if item > 30))

print("\n---- Prints how many invalid entries were removed during the cleaning process. ----")
print(len(ages) - len(ages_valid))
```

```python
products = ["Laptop","Phone","Tablet","Monitor","Headphones"] 
amazon_prices = [75000, 45000, 30000, 20000, 5000] 
flipkart_prices = [72000, 47000, 31000, 19500, 4800]

products_combined = []
for i, item in enumerate(products):
    cheaper_platform = ''
    diff = 0
    if amazon_prices[i] < flipkart_prices[i]:
        cheaper_platform = 'Amazon'
        diff = flipkart_prices[i] - amazon_prices[i]
    else:
        cheaper_platform = 'Flipkart'
        diff = amazon_prices[i] - flipkart_prices[i]
    products_combined.append({
        "Product_Name" : item,
        "Amazon_Price" : amazon_prices[i],
        "Flipkart_Price" : flipkart_prices[i],
        "Cheaper_platform" : cheaper_platform,
        "Difference" : diff
    }        
    )

print(f"| {'------------':<12} | {'----------------':<16} | {'----------':<10} |")
print(f"| {'Product_Name':<12} | {'Cheaper_Platform':<16} | {'Difference':<10} |")
print(f"| {'------------':<12} | {'----------------':<16} | {'----------':<10} |")
for item in products_combined:
    print(f"| {item["Product_Name"]:<12} | {item["Cheaper_platform"]:<16} | {item["Difference"]:<10} |")
print(f"| {'------------':<12} | {'----------------':<16} | {'----------':<10} |")
```

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
print("---- Separates login hours into even-hour logins and odd-hour logins. ----")
print("Odd hours : ", list(item for item in login_hours if item % 2 != 0))
print("Even hours : ", list(item for item in login_hours if item % 2 == 0))

print("\n---- Identifies login hours that fall into peak activity time (hours greater than 5). ----")
print("Peak activity hours : ", list(item for item in login_hours if item > 5))

print("\n---- Sorts the login hours in ascending order. ----")
print("Activity hours sorted: ", sorted(login_hours))

print("\n---- Calculates how many logins occurred during peak hours. ----")
print(len(list(item for item in login_hours if item > 5)))
```

```python
orders = [ ["ORD001", "Laptop",   75000, "Completed"], 
           ["ORD002", "Mouse",      500, "Completed"], 
           ["ORD003", "Laptop",   75000, "Returned"], 
           ["ORD004", "Keyboard",  1500, "Completed"], 
           ["ORD005", "Laptop",   75000, "Completed"], 
           ["ORD006", "Monitor",  20000, "Fraud"], 
           ["ORD007", "Mouse",      500, "Completed"], 
           ["ORD008", "Laptop",   75000, "Completed"] ]

print("---- Display All Orders Using a nested loop, print the entire order table row by row so that each order’s details are displayed clearly. ----")
print(f"|{"--------":<8} | {"------------":<12} | {"--------":8} | {"---------":<9} |")
print(f"|{"Order_ID":<8} | {"Product_Name":<12} | {"   Price":8} | {"Status":<9} |")
print(f"|{"--------":<8} | {"------------":<12} | {"--------":8} | {"---------":<9} |")
for row in orders:   
    print(f"|{row[0]:<8} | {row[1]:<12} | {row[2]:8.2f} | {row[3]:<9} |")
print(f"|{"--------":<8} | {"------------":<12} | {"--------":8} | {"---------":<9} |")

print("\n---- Detect Suspicious Activity The system suspects fraud when: the same product appears more than 3 times, OR an order has status 'Fraud' ----")
print("-- Write a program that scans the dataset and identifies suspicious products or orders. You must use nested loops to compare product occurrences. --")
suspect = []
unique_products = []
for row in orders:
    unique_products.append(row[1])
unique_products = list(set(unique_products))
unique_product_count = [0] * len(unique_products)
for row in orders:
    index = unique_products.index(row[1]) 
    unique_product_count[index] = unique_product_count[index] + 1

for row in orders:    
    for i, item in enumerate(row):
        match i:
            case 1:
                index = unique_products.index(item) 
                if unique_product_count[index] > 3:
                    suspect.append(row)
            case 3:
                if item == "Fraud":
                    suspect.append(row)

print("Suspicious orders")
print(f"|{"--------":<8} | {"------------":<12} | {"--------":8} | {"---------":<9} |")
print(f"|{"Order_ID":<8} | {"Product_Name":<12} | {"   Price":8} | {"Status":<9} |")
print(f"|{"--------":<8} | {"------------":<12} | {"--------":8} | {"---------":<9} |")
for row in suspect:   
    print(f"|{row[0]:<8} | {row[1]:<12} | {row[2]:8.2f} | {row[3]:<9} |")
print(f"|{"--------":<8} | {"------------":<12} | {"--------":8} | {"---------":<9} |")

print("\n---- Total revenue generated for the day ----")
total_revenue = 0
for row in  orders:
    if row[3] == "Completed":
        total_revenue += row[2]
print("Total revenue : ", total_revenue)

print("\n---- Count how many orders fall into each category ----")
catagory_count = {}
for row in orders:
    if row[3] in catagory_count:
        catagory_count[row[3]] += 1        
    else:
        catagory_count[row[3]] = 1    
completed_count = catagory_count["Completed"]
returned_count = catagory_count["Returned"]
Fraud_count = catagory_count["Fraud"]
print("Completed : ", completed_count)
print("Returned : ", returned_count)
print("Fraud : ", Fraud_count)

print("\n---- Generate a List of Unique Products ----") 
unique_products = []
for row in orders:
    if row[1] in unique_products:
        continue
    unique_products.append(row[1])
print("Unique products : ", unique_products)

print("\n---- Product Sales Analysis For each unique product ----") 
product_count = [0] * len(unique_products)
product_reveue = [0] * len(unique_products)
for row in orders:
    if row[3] != "Completed":
        continue
    index = unique_products.index(row[1]);
    product_count[index] += 1
    product_reveue[index] += row[2]
for i, item in enumerate(unique_products):
    print(f"{item:<10} : No of Sales -  {product_count[i]:<2}, Revenue - {product_reveue[i]:9.2f}")
    
max_sold_count = max(product_count) 
print("\n---- Most Sold Product Based on the frequency of orders  ----") 
print("Product name : ", unique_products[product_count.index(max_sold_count)], ", Count : ", max_sold_count)

print("\n---- Remove Fraud Orders Create a cleaned dataset by removing orders whose status is 'Fraud'.  ----") 
clean_orders = list(item for item in orders if item[3] != 'Fraud')
print(f"|{"--------":<8} | {"------------":<12} | {"--------":8} | {"---------":<9} |")
print(f"|{"Order_ID":<8} | {"Product_Name":<12} | {"   Price":8} | {"Status":<9} |")
print(f"|{"--------":<8} | {"------------":<12} | {"--------":8} | {"---------":<9} |")
for row in clean_orders:   
    print(f"|{row[0]:<8} | {row[1]:<12} | {row[2]:8.2f} | {row[3]:<9} |")
print(f"|{"--------":<8} | {"------------":<12} | {"--------":8} | {"---------":<9} |")
```
